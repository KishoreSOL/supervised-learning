```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.linear_model import LinearRegression

# Generate random data
np.random.seed(0)
X1 = np.random.rand(100)
X2 = np.random.rand(100)
y = 2 * X1 + 3 * X2 + np.random.rand(100) * 0.2  # Simulated linear relationship with noise

# Create a 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Fit a multiple linear regression model
X = np.column_stack((X1, X2))
model = LinearRegression()
model.fit(X, y)

# Generate predictions for the entire input space
X1_range = np.linspace(min(X1), max(X1), 20)
X2_range = np.linspace(min(X2), max(X2), 20)
X1_grid, X2_grid = np.meshgrid(X1_range, X2_range)
X_grid = np.column_stack((X1_grid.ravel(), X2_grid.ravel()))
y_pred = model.predict(X_grid)

# Plot the data points
ax.scatter(X1, X2, y, c='b', marker='o', label='Data Points')

# Plot the regression plane
ax.plot_surface(X1_grid, X2_grid, y_pred.reshape(X1_grid.shape), alpha=0.5, cmap='viridis', label='Regression Plane')

ax.set_xlabel('X1')
ax.set_ylabel('X2')
ax.set_zlabel('Y')
ax.set_title('Multiple Linear Regression')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.linear_model import LinearRegression

# Generate random data
np.random.seed(0)
X1 = np.random.rand(100)
X2 = np.random.rand(100)
X3 = np.random.rand(100)
y = 2 * X1 + 3 * X2 + 4 * X3 + np.random.rand(100) * 0.2  # Simulated linear relationship with noise

# Create a 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Fit a multiple linear regression model
X = np.column_stack((X1, X2, X3))
model = LinearRegression()
model.fit(X, y)

# Generate predictions for the entire input space
X1_range = np.linspace(min(X1), max(X1), 20)
X2_range = np.linspace(min(X2), max(X2), 20)
X3_range = np.linspace(min(X3), max(X3), 20)
X1_grid, X2_grid, X3_grid = np.meshgrid(X1_range, X2_range, X3_range)
X_grid = np.column_stack((X1_grid.ravel(), X2_grid.ravel(), X3_grid.ravel()))
y_pred = model.predict(X_grid)

# Plot the data points
ax.scatter(X1, X2, X3, c=y, cmap='viridis', marker='o', label='Data Points')

# Plot the regression plane
ax.scatter(X1_grid, X2_grid, X3_grid, c=y_pred, cmap='viridis', marker='.', label='Predicted Values')

ax.set_xlabel('X1')
ax.set_ylabel('X2')
ax.set_zlabel('X3')
ax.set_title('Multiple Linear Regression')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_regression
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.inspection import plot_partial_dependence

# Generate random data with four independent variables
X, y = make_regression(n_samples=100, n_features=4, noise=0.2, random_state=0)

# Fit a multiple linear regression model
model = GradientBoostingRegressor(random_state=0)
model.fit(X, y)

# Create partial dependence plots
fig, ax = plt.subplots(figsize=(12, 6))
plot_partial_dependence(model, X, features=[0, 1, 2, 3], grid_resolution=50, ax=ax)
plt.suptitle('Partial Dependence Plots for Multiple Regression')
plt.subplots_adjust(top=0.9)
plt.show()
```
